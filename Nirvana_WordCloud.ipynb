# Introduction
Wordclouds are a fun and visually appealing way of viewing frequently used words in a dataset. An example of when one of these could be useful in a business setting is identifying common themes in customer feedback surveys.

As a big fan of Nirvana, I decided it could be fun to see some of the most common words used in thier lyrics.



```python
#Connecting the necessary libraries for this project
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import os
```

```python
#Grabbing the Nirvana lyrics file I downloaded from Kaggle and turning it to a data frame
csv_path = '/content/Nirvana_lyrics.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
text = " ".join(review for review in df.lyrics.astype(str))
print("There are {} words in the comination of all cells in the column lyrics".format(len(text)))
```

I'll set the stopwords below, if there are any words that appear in the wordcloud that I would like to remove for the project. I typically do a first run to see what the cloud looks like, then remove any filler or irrelevant words from there.

```python
#set the stopwords, words we don't want to include in the wordcloud
stopwords = set(STOPWORDS)
stopwords.update(["Nirvana","Rape","'"])
```

Create the wordcloud visualization

```python
wordcloud = WordCloud( width=400,
                       height=330,
                       max_words=100,
                       colormap='brg',
                       stopwords=stopwords,
                       background_color="White").generate(text)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```

I'm interested in seeing how the lyrics from 80s differ from the lyrics in the 90s, let's break this into 2 different wordclouds.

```python
df2 = df.loc[(df['Release Year']<1990)]
df2.head()
```

```python
text2 = " ".join(review for review in df2.lyrics.astype(str))
```

```python
wordcloud = WordCloud( width=400,
                       height=330,
                       max_words=100,
                       colormap='brg',
                       stopwords=stopwords,
                       background_color="white").generate(text2)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```

```python
df3 = df.loc[(df['Release Year'].between(1990,1999))]
df3.head()
```

```python
text3 = " ".join(review for review in df3.lyrics.astype(str))
```

```python
wordcloud = WordCloud( width=400,
                       height=330,
                       max_words=100,
                       colormap='brg',
                       stopwords=stopwords,
                       background_color="white").generate(text3)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```

#Conclusion
It's interesting to view the overall themes of Nirvana lyrics as well as the differentiation between the lyrics of songs released in 2 different decades.

The lyrics of the songs in the 80s have a lot of adolescent themes to them, with words like "Little Girl", "Recess", and "Teacher". The 90s lyrics feature a lot more repetative phrases  "Yeah Yeah", "Hello Hello", and "Dive Dive" and possibly a focus on wisdom and experience with how often words like "know", "grandma", and "think".
